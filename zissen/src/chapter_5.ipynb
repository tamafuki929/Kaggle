```python
import pandas as pd
path = "../dataset/chapter_5/"
cus = pd.read_csv(path + "customer_join.csv")
ulog_mon = pd.read_csv(path + "use_log_months.csv")
ulog_mon.head()
```

## 退会予測(決定木による分類)
 当月のデータと前月のデータから来月の退会or stayを分類

```python
years_months = list(ulog_mon["年月"].unique())
ulog = pd.DataFrame()
for i in range(1, len(years_months)):
    tmp = ulog_mon.loc[ulog_mon["年月"] == years_months[i]].copy()
    tmp.rename(columns = {"count":"count_0"}, inplace = True)
    tmp_before = ulog_mon.loc[ulog_mon["年月"] == years_months[i]].copy()
    del tmp_before["年月"]
    tmp_before.rename(columns = {"count":"count_1"}, inplace = True)
    tmp = pd.merge(tmp, tmp_before, on = "customer_id", how = "left")
    ulog = pd.concat([ulog, tmp], ignore_index = True)
ulog.head()
```

### 退会顧客のデータを作成
・退会顧客は退会した月(end_date)の一ヶ月前に退会を決意する.  
→ "年月"を退会を退会を決意した月と考える.  
・退会顧客の退会決意当月(end_date - 1), 退会決意前月(end_date - 2)のデータを用いる

```python
from dateutil.relativedelta import relativedelta
exit_cus = cus.loc[cus["is_deleted"] == 1].copy()
exit_cus["end_date"] = pd.to_datetime(exit_cus["end_date"])
exit_date = []
for i in range(len(exit_cus)):
    exit_date.append(exit_cus["end_date"].copy().iloc[i] - relativedelta(months = 1))
exit_cus["exit_date"] = exit_date
exit_cus["年月"]  = exit_cus["exit_date"].dt.strftime("%Y%m")
ulog["年月"] = ulog["年月"].astype(str)
exit_ulog = pd.merge(ulog, exit_cus, on = ["customer_id", "年月"], how = "left")
exit_ulog = exit_ulog.dropna(subset = ["name"])

exit_ulog.head()
```

## 継続顧客のデータを作成

```python
conti = cus.loc[cus["is_deleted"] == 0].copy()
conti_ulog = pd.merge(ulog, conti, on = ["customer_id"], how = "left")
conti_ulog = conti_ulog.dropna(subset = ["name"])
conti_ulog.info()
```

### アンダーサンプリング
・退会と継続が不均衡  
・顧客あたりの継続データが一つになるようにダウンサンプリング

```python
conti_ulog = conti_ulog.sample(frac = 1).reset_index(drop = True) # frac = 1でシャッフル
conti_ulog = conti_ulog.drop_duplicates(subset = "customer_id") # 上側優先で重複削除
conti_ulog.info()
```

```python
predict_data = pd.concat([conti_ulog, exit_ulog], ignore_index = True)
predict_data.head()
```

```python
predict_data["now_date"] = pd.to_datetime(predict_data["年月"], format = "%Y%m")
predict_data["start_date"] = pd.to_datetime(predict_data["start_date"])
period = []
for i in range(len(predict_data)):
    tmp = predict_data.iloc[i, :].copy()
    delta = relativedelta(tmp["now_date"], tmp["start_date"])
    period.append(int(delta.years * 12 + delta.months))
predict_data["period"] = period
predict_data.isnull().sum()
```

## 予測に用いる変数の選択, ダミー変数化

```python
target_col = ["gender", "count_1", "routine_flg", "period", "is_deleted"]
predict_data = predict_data[target_col]
predict_data.head()
```

```python
predict_data = pd.get_dummies(predict_data)
predict_data.head()
```

```python
# 不要な変数列を削除 (例: 女性でなければ男性なので, genderはどちらかだけで良い)
drop_col = ["gender_M"]
predict_data = predict_data.drop(drop_col, axis = 1)
predict_data.head()
```

### 決定木を学習

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

exit = predict_data.loc[predict_data["is_deleted"] == 1]
conti = predict_data.loc[predict_data["is_deleted"] == 0].sample(len(exit))
X = pd.concat([exit, conti], ignore_index = True)
y = X["is_deleted"]
del X["is_deleted"]
X_train, X_test, y_train, y_test = train_test_split(X, y)

model = DecisionTreeClassifier(random_state = 0, max_depth = 5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
results = pd.DataFrame({"y_test":y_test, "y_pred":y_pred})
results.head()
```

```python
print(model.score(X_test, y_test))
print(model.score(X_train, y_train))
```
