```python
import pandas as pd
import matplotlib.pyplot as plt
path = '../dataset/chapter_2/'
```

```python
ur = pd.read_csv(path + 'uriage.csv')
ur.head()
```

```python
kokyaku = pd.read_excel(path+'kokyaku_daicho.xlsx')
kokyaku.head()
```

## データの揺れを補正

```python
print(ur.shape)
ur["item_name"] = ur["item_name"].str.upper() # 大文字に
ur["item_name"] = ur["item_name"].str.replace("　", "") # 全角スペース削除
ur["item_name"] = ur["item_name"].str.replace(" ", "") # 半角スペース削除
# byで指定した列基準にソート , 複数指定すれば後ろから行う, ascendingsはTrueで昇順
ur.sort_values(by = ["item_name"], ascending = True)
```

```python
pd.unique(ur["item_name"])
```

```python
ur["purchase_date"] = pd.to_datetime(ur["purchase_date"])
ur["purchase_month"] = ur["purchase_date"].dt.strftime("%Y%m")
ur.isnull().any(axis = 0)
```

## 欠損値補完

```python
flg_is_null = ur["item_price"].isnull()
# nulを含むitemの種類ごと
for trg in list(ur.loc[flg_is_null, "item_name"].unique()):
    # item-xの欠損値は, 他のitem-xの購買データの値を代入
    price = ur.loc[(~flg_is_null) & (ur["item_name"] == trg), "item_price"].copy().max()
    ur["item_price"].loc[(flg_is_null) & (ur["item_name"] == trg)] = price
ur.isnull().any(axis = 0)
```

```python
kokyaku.head()
```

```python
kokyaku["顧客名"] = kokyaku["顧客名"].str.replace("　", "")
kokyaku["顧客名"] = kokyaku["顧客名"].str.replace(" ", "")
kokyaku.head()
```

## 日付の揺れを補正(フォーマットの統一)

```python
# シリアル値: excelにおける１９００年１月1日を起点にカウントした時間を整数値で表した値
# シリアル値を日付に
flg_is_serial = kokyaku["登録日"].astype("str").str.isdigit()
Serial = pd.to_timedelta(kokyaku.loc[flg_is_serial, "登録日"].astype("float"), 
                           unit = "D")  + pd.to_datetime("1900/01/01") # unitは単位, D:date
String = pd.to_datetime(kokyaku.loc[~flg_is_serial, "登録日"])
kokyaku["登録日"] = pd.concat([Serial, String])
kokyaku
```

```python
kokyaku["登録年月"] = kokyaku["登録日"].dt.strftime("%Y%m")
join_data = pd.merge(ur, kokyaku, left_on = "customer_name", right_on = "顧客名", 
                    how = "left")
join_data = join_data.drop("customer_name", axis = 1)
join_data
```
